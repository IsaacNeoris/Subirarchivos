```python
tiempo=data2[['Year','Season','Medal','NOC']].dropna()
```

```python
tiempo #ACT 17: Describir un ejemplo donde se utilizaría matriz de correlación y como ayudaría
```

```python
df_2= pd.get_dummies(tiempo[['Medal']]) #Transformar una columna en varios con 0s y 1s
df_2.head()
```

```python
result= pd.concat([tiempo,df_2], axis=1) #Concatenar df1 con df2 ([df1,df2], axis)
result.head()
```

```python
tiempo2= result.groupby(['Year','Medal','Season']).sum()
tiempo2.head(6)
```

```python
corr=tiempo2.corr() #Correlación con coeficientes de Pearson
corr.style.background_gradient(cmap='coolwarm')
```
